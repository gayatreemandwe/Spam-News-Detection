# NLP

## Clean

```python
import nltk
```

```python
import re
```

```python
from nltk.corpus import stopwords
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
ps = WordNetLemmatizer()
```

```python
stopwords = stopwords.words('english')
```

```python
dataset['text']
```

```python
nltk.download('wordnet')
```

```python
def clean_row(row):
    row = row.lower()
    row = re.sub('[^a-zA-Z]',' ',row)
    token = row.split()
    text=(ps.lemmatize(word)for word in token if not word in stopwords)
    cleanned_news = ' '.join(text)
    return cleanned_news
```

```python
dataset['text']
```

```python
dataset['text'] = dataset['text'].apply(lambda x :clean_row(x))
```

```python
dataset['text']
```

## encoded

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
import pandas as pd

# Load your data
x = dataset.iloc[:35000, 0]
y = dataset.iloc[:35000, 1]

# Split
train_data, test_data, train_label, test_label = train_test_split(x, y, test_size=0.2, random_state=0)

# Vectorizer
vectorizer = TfidfVectorizer(max_features=50000, lowercase=False, ngram_range=(1, 2))

# Fit only on train
vec_train = vectorizer.fit_transform(train_data)
vec_test = vectorizer.transform(test_data)

# Debug print to verify shape
print("Train shape:", vec_train.shape)
print("Test shape:", vec_test.shape)
print("Feature names length:", len(vectorizer.get_feature_names_out()))

# Convert to DataFrames
training_data = pd.DataFrame(vec_train.toarray(), columns=vectorizer.get_feature_names_out())
testing_data = pd.DataFrame(vec_test.toarray(), columns=vectorizer.get_feature_names_out())
```
